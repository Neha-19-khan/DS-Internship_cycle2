```python
#Load CNN/DailyMail Dataset from HuggingFace
from datasets import load_dataset

# Load a small sample of the dataset
dataset = load_dataset("cnn_dailymail", "3.0.0", split='train[:100]')
print("Sample loaded:")
print(dataset[0])
```

```python
!python -m spacy download en_core_web_sm
```

```python
#Text Preprocessing (for Extractive Summarization)
import spacy

nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    doc = nlp(text)
    sentences = [sent.text.strip() for sent in doc.sents]
    return sentences

# Example
example_article = dataset[0]['article']
sentences = preprocess_text(example_article)
print(f"Total Sentences: {len(sentences)}")
print(sentences[:5])
```

```python
# Extractive Summarization using spaCy (Top-N Sentences)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def extractive_summary(text, num_sentences=3):
    sentences = preprocess_text(text)
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(sentences)
    similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

    sentence_scores = similarity_matrix.sum(axis=1)
    ranked_sentences = [sentences[i] for i in np.argsort(sentence_scores)[-num_sentences:]]
    summary = " ".join(ranked_sentences)
    return summary

print("Extractive Summary:")
print(extractive_summary(example_article))
```

```python
from transformers import pipeline

# Load summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Shorten article due to token limits
short_text = example_article[:1024]
summary = summarizer(short_text, max_length=130, min_length=30, do_sample=False)

print("Abstractive Summary:")
print(summary[0]['summary_text'])
```

###  Task 2: Text Summarization  Insight

**Objective:**  
To build a system that can summarize long news articles into short, meaningful summaries.

**Approach:**  
-  *Extractive Summarization* using **TF-IDF + Cosine Similarity** to pick key sentences.
-  *Abstractive Summarization* using **BART (facebook/bart-large-cnn)** model via HuggingFace.

**Results:**  
- Extractive summaries were grammatically correct but lacked fluency.
- Abstractive summaries were more human-like and fluent.

**Challenges Faced:**  
- HuggingFace models require trimming text to max 1024 tokens.
- TensorFlow backend caused errors  solved by switching to **PyTorch** (`pip install torch`).
- `spaCy` model (`en_core_web_sm`) was missing initially  installed manually.

**Key Insight:**  
Abstractive summarization gives better quality outputs but needs more resources. For real-world apps like news apps or email summarizers, abstractive models are preferred.
